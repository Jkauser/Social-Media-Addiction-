# 1) Test to check relation between Gender and how often used social media account.

#### Null Hypothesis: Gender and used of social media are independent.

#### Alternate Hypothesis: Gender and used of social media are dependent.

```python
#contingency table
cross1=pd.crosstab(data.Gender,data.Q3,margins=True)
cross1
```

```python
#Chi-square test
from scipy.stats import chi2_contingency
D1=[[56,4,2,1,19],[110,7,3,3,14],[2,0,0,0,0]]
stat,p ,dof,expected= chi2_contingency(D1)
alpha=0.05
print("p value is" + str(p))
if p<= alpha:
    print('Reject Null Hypothesis')
else:
    print('Fail to reject H0 i.e Accept Null Hypothesis')
```

#### Here,p-value=0.46 is greater than 0.05. Therefore we may fail to regect Null Hypothesis.

#### Interpretation: Gender and used of social media may be independent
